##7. Splitting the dataset into train and test data

```python
from sklearn.model_selection import train_test_split
X = data.drop("price",axis=1)
y = data['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)
```

# performing mutual info regression(measures the dependency between the variables)

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info = mutual_info_regression(X_train.fillna(0), y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectPercentile
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=40)
selected_top_columns.fit(X_train.fillna(0), y_train)
selected_top_columns.get_support()
```

#Selected features(selected top columns)

```python
X_train.columns[selected_top_columns.get_support()]
```

```python
X_train=X_train[X_train.columns[selected_top_columns.get_support()]]
```

```python
X_train
```

```python
X_test = X_test[X_test.columns[selected_top_columns.get_support()]]
```

```python
X_test
```

##8. Model Implementation

#Ridge Regression

```python
from sklearn.linear_model import Ridge, Lasso
model = Ridge(alpha=10.0)

model.fit(X_train, y_train)

model.score(X_test,y_test)
```

```python
model1 = Ridge(alpha=1000.0)

model1.fit(X_train, y_train)

model1.score(X_test,y_test)
```

```python
model2 = Ridge(alpha=10000.0)

model2.fit(X_train, y_train)

train_score_m2 =model2.score(X_train,y_train)
test_score_m2=model2.score(X_test,y_test)

print("The train score for ls model is {}".format(train_score_m2))
print("The test score for ls model is {}".format(test_score_m2))
```

Ridge regression gives 84.3% accuracy with test dataset

##When the alpha value is greater than 10^5, the accuracy reduces
#model2 i.e., alpha = 10000 gives the maximum accuracy

```python
model3 = Ridge(alpha=100000.0)

model3.fit(X_train, y_train)

model3.score(X_test,y_test)
```

#Lasso Regression

```python
#Lasso regression model
print("\nLasso Model............................................\n")
lasso = Lasso(alpha = 1000)
lasso.fit(X_train,y_train)
train_score_ls =lasso.score(X_train,y_train)
test_score_ls =lasso.score(X_test,y_test)

print("The train score for ls model is {}".format(train_score_ls))
print("The test score for ls model is {}".format(test_score_ls))
```

#Lasso regression predicts with the accuracy of 83.6% with test dataset
